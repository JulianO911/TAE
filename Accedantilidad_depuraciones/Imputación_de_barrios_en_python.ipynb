```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
accidentes = pd.read_csv(io.BytesIO(uploaded['accidentes.csv']),encoding="latin-1")
accidentes
```

```python
accidentesBarrio = accidentes[~accidentes["BARRIO"].isnull()]
accidentesBarrio = accidentesBarrio.drop(["Unnamed: 0","CBML","DIRECCION.ENCASILLADA","DISE√ëO","EXPEDIENTE","LOCATION"],axis=1)
accidentesBarrio = accidentesBarrio.dropna()
accidentesBarrio
```

```python
X = accidentesBarrio[["A√ëO","CLASE_ACCIDENTE","COMUNA","HORA","X","Y"]]
y = accidentesBarrio["BARRIO"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,train_size=0.7)
X_train
```

```python
y_train = y_train.astype("category")
y_train
```

```python
from sklearn import preprocessing
def convert(Data):
  for column in Data.columns:
    number = preprocessing.LabelEncoder()
    if column != "NRO_RADICADO":
      Data[column] = number.fit_transform(Data[column])
  return Data
```

```python
X_train = convert(X_train)
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train) 
X_train
```

```python
X_test = convert(X_test)
scale = StandardScaler()
X_test = scale.fit_transform(X_test) 
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
accidentesBarrioNull = accidentes[accidentes["BARRIO"].isnull()]
accidentesBarrioNull = accidentesBarrioNull.drop(["Unnamed: 0","CBML","DIRECCION.ENCASILLADA","DISE√ëO","EXPEDIENTE","LOCATION"],axis=1)
accidentesBarrioNull = accidentesBarrioNull[["A√ëO","CLASE_ACCIDENTE","COMUNA","HORA","X","Y"]]
accidentesBarrioNull = convert(accidentesBarrioNull)
scale = StandardScaler()
accidentesBarrioNull = scale.fit_transform(accidentesBarrioNull) 
accidentesBarrioNull
```

```python
accidentesBarrioNull
```

```python
barriosPred = knn.predict(accidentesBarrioNull)
```

```python
accidentes.loc[accidentes["BARRIO"].isnull(),"BARRIO"] = barriosPred
```
