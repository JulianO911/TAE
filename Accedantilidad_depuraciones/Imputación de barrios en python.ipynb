```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
accidentes = pd.read_csv(io.BytesIO(uploaded['accidentes.csv']),encoding="latin-1")
accidentes
```

```python
accidentesBarrio = accidentes[~accidentes["BARRIO"].isnull()]
accidentesBarrio
```

```python
X = accidentesBarrio[["X","Y"]]
y = accidentesBarrio["BARRIO"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,train_size=0.7)
X_train
```

```python
y_train = y_train.astype("category")
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 500)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
accidentesBarrioNull = accidentes[accidentes["BARRIO"].isnull()]
accidentesBarrioNull
```

```python
barriosPred = knn.predict(accidentesBarrioNull[["X","Y"]])
```

```python
barriosPred
```

```python
accidentes.loc[accidentes["BARRIO"].isnull(),"BARRIO"] = barriosPred
```

```python
accidentes[accidentes["BARRIO"].isnull()]
```

```python
accidentes["BARRIO"] = accidentes.BARRIO.astype(str)
accidentesNuevos = accidentes.iloc[:,1:]
accidentesNuevos
```
